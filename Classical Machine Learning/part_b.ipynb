## Voting Dataset

```python
%%capture
lambdas = [0,0.001,0.01,0.1]

auc_vals = []

for lambda_ in lambdas:
    auc_vals.append(logreg(data_path=base+"voting",lambda_=lambda_,use_cross_validation=True)[6])
```
